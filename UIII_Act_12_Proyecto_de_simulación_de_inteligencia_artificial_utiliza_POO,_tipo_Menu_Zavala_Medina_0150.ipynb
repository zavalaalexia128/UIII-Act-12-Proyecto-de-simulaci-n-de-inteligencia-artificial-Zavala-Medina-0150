##### Problema Pescador

##### Pescador: Embarcación, zona de pesca, especies objetivo, redes utilizadas, licencia de pesca y puerto base.

```python
import random

class Pescador:
    def __init__(self, nombre, embarcacion, zona_pesca, especies_objetivo, redes_utilizadas, licencia_pesca, puerto_base):
        # Atributos de la clase
        self.nombre = nombre
        self.embarcacion = embarcacion
        self.zona_pesca = zona_pesca
        self.especies_objetivo = especies_objetivo
        self.redes_utilizadas = redes_utilizadas
        self.licencia_pesca = licencia_pesca
        self.puerto_base = puerto_base

    def simular_ia_pesca(self):
        # Simula una IA basica (Sistema Experto) para predecir el exito de la pesca
        # Clima aleatorio simulado por el entorno
        clima = random.choice(["Favorable", "Tormentoso", "Caluroso"])

        # Logica de la IA basada en reglas
        if self.licencia_pesca.lower() != "activa":
            return "ALERTA IA: Pesca suspendida. Licencia inactiva o vencida."

        if clima == "Tormentoso":
            return f"RECOMENDACION IA: Clima peligroso en {self.zona_pesca}. Regresar a {self.puerto_base}."

        if self.redes_utilizadas.lower() == "malla fina" and "Tiburón" in self.especies_objetivo:
            return "SUGERENCIA IA: Red inadecuada. El Tiburon rompera la malla fina. Cambiar red."

        # Si pasa los filtros, calcula una probabilidad de exito
        probabilidad = random.randint(50, 95)
        return f"PREDICCION IA: Exito del {probabilidad}% en {self.zona_pesca} buscando {self.especies_objetivo}."

    def mostrar_info_tabla(self):
        # Devuelve los datos formateados para alinearse en la tabla
        return f"| {self.nombre:<12} | {self.embarcacion:<12} | {self.zona_pesca:<13} | {self.especies_objetivo:<18} | {self.redes_utilizadas:<12} | {self.licencia_pesca:<8} | {self.puerto_base:<12} |"


# --- CONFIGURACION INICIAL DE DATOS ---
# Lista para almacenar a los pescadores (Objetos)
flota = [
    Pescador("Carlos", "Titan", "Pacifico Norte", "Atun, Dorado", "Arrastre", "Activa", "Ensenada"),
    Pescador("Marina", "La Nina", "Golfo de Mex.", "Camaron", "Malla fina", "Activa", "Veracruz"),
    Pescador("Pedro", "Tiburon II", "Mar de Cortes", "Tiburon, Raya", "Malla fina", "Activa", "Guaymas")
]

# --- MENU INTERACTIVO ---
opcion = ""
while opcion != "4":
    print("\n" + "="*50)
    print(" SIMULADOR DE PESCA INTELIGENTE (IA Y POO) ")
    print("="*50)
    print("1. Ver lista de pescadores (Tabla)")
    print("2. Agregar nuevo pescador")
    print("3. Ejecutar Prediccion con IA")
    print("4. Salir")
    print("="*50)

    opcion = input("Selecciona una opcion (1-4): ")

    if opcion == "1":
        # Mostrar el listado en forma de tabla bien distribuida
        print("\n" + "-" * 105)
        print(f"| {'Nombre':<12} | {'Embarcacion':<12} | {'Zona Pesca':<13} | {'Especies Obj.':<18} | {'Redes':<12} | {'Licencia':<8} | {'Puerto Base':<12} |")
        print("-" * 105)
        for p in flota:
            print(p.mostrar_info_tabla())
        print("-" * 105)

    elif opcion == "2":
        # Entrada de datos interactiva
        print("\n--- REGISTRAR NUEVO PESCADOR ---")
        nombre = input("Nombre del pescador: ")
        embarcacion = input("Nombre de la embarcacion: ")
        zona = input("Zona de pesca: ")
        especies = input("Especies objetivo (ej. Merluza, Pulpo): ")
        redes = input("Tipo de redes utilizadas: ")
        licencia = input("Estado de la licencia (Activa / Vencida): ")
        puerto = input("Puerto base: ")

        # Creamos el nuevo objeto y lo guardamos en la lista
        nuevo_pescador = Pescador(nombre, embarcacion, zona, especies, redes, licencia, puerto)
        flota.append(nuevo_pescador)
        print(f"\n¡{nombre} ha sido anadido exitosamente a la flota!")

    elif opcion == "3":
        # Ejecutar la funcion de simulacion IA
        print("\n--- EJECUTANDO ANALISIS DE IA ---")
        if len(flota) == 0:
            print("No hay pescadores registrados en la flota.")
        else:
            for p in flota:
                print(f"\n> Pescador: {p.nombre} ({p.embarcacion})")
                # Llamada a la funcion inteligente
                resultado_ia = p.simular_ia_pesca()
                print(resultado_ia)

    elif opcion == "4":
        print("\nGracias por usar el Simulador IA de Pesca. Fin del programa.")
        print("Programa realizado por Zavala Medina NL = 59")

    else:
        print("\nOpcion no valida. Por favor, elige un numero del 1 al 4.")
```
